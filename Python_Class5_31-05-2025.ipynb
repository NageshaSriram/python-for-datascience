```python
# add new column if say Yes if the name start with vowel otherwise no
start_with_vowel = []
for i in df.name:
  if i[0] in 'aeiouAEIOU': # give it in a tuple ('a', 'e', ...)
    start_with_vowel.append('Yes')
  else:
    start_with_vowel.append('No')
df['start_with_vowel'] = start_with_vowel
```

```python
df
```

```python
def vowel(x):
  if x[0] in 'aeiouAEIOU':
    return 'Yes'
  else:
    return 'No'
```

```python
df['start_with_vowel'] = df.name.apply(vowel)
```

```python
df
```

```python
# Create a function to find length of a name column and add it to new column
def find_length(name):
  return len(name)

df['name_length'] = df.name.apply(find_length)
df
```

```python
# add age number digits and added it to new column
def add_digit(n):
  digits = [int(x) for x in str(n)]
  return sum(digits)

df['added_age_digits'] = df.age.apply(add_digit)
df
```

```python
df['New_added_age_digits'] = df.age.apply(lambda x: sum(map(int, str(x))))
```

# Join
- common column

# Left Join A - B
- all data from table A(left) and matching data from table B(right)

# Right Join A - B
- all data from table B(right) and matching data from table A(left)

# Inner Join

- common in the ID's

# Outer join

- all the data from both the table we will get

## 1. Left Join(A + B) = Right Join(B + A)
## 2. No of rows A != no of rows of Inner Join (A+B)
## 3. No of rows A != No of rows of Left Join (A + B) : this statement is True only if Keys are unique
